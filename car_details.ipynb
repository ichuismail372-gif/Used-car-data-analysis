# Data Cleaning

# Clean numerical columns with charecters

```python
#Remove text unit and convert to numbers
df['engine']=pd.to_numeric(
    df['engine']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True)
    .replace('',None),
    errors='coerce'
    )

df.head(5)
```

```python
#Remove text unit and convert to numbers
df['max_power']=pd.to_numeric(
    df['max_power']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True)
    .replace('',None),
    errors='coerce'
    )

df.head(5)
```

```python
#Remove text unit and convert to numbers
df['mileage']=pd.to_numeric(
    df['mileage']
    .astype(str)
    .str.replace(r'[^0-9,-]','',regex=True)
    .replace('',None),
    errors='coerce'
    )

df.head(5)
```
